# Problem 5

When n = 1,

$
\begin{equation}
2 \mu \dot{\epsilon} = \rho g w \\
2 \mu \frac{\dot{w}}{L} = \rho gw \\
\dot{w} = \frac{L \rho g}{2 \mu}
\end{equation}
$

The solution is,
$w(t) = w(0) \exp{\frac{L \rho g}{2 \mu}t} \: $

When $n \neq 1$,

$
\begin{equation}
\big( \frac{1}{B} \dot{\epsilon} \big)^{\frac{1}{n}} = \rho g w \\
\big( \frac{\dot{w}}{BL} \big)^{\frac{1}{n}} = \rho g w \\
\dot{w} = \rho^{n} g^{n} w^{n} BL
\end{equation}
$

The solution for $n = 2$ is given as 
$w(t) = - \frac{2}{t^2 + w(0)}$

The solution for $n = 3$ is given as 
$w(t) = - \frac{1}{\sqrt{w(0) - t^2}}$
