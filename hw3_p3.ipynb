```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
v = 5.0 * 0.01 / 3.1536e7 # 5 cm/yr to 5 m/sec.
rho = 3300.0 # kg/m^3
cp = 1000.0 # J/kg/K
k = 3.0 #
T1 = 1300.0 # K
T0 = 0.0

time = np.linspace(0, 130, 131)
```

```python
# Data from the table
q_data = np.array([47.31, 65.73, 56.94, 100.48, 176.26, 166.22, 172.50, 119.32, 38.52, 239.48, 267.96])
t_data = np.array([130, 95, 85, 17.5, 7.5, 7.5, 7.5, 15, 50, 5.25, 3.5])
```

```python
L = 100e3
R = 0.5 * rho * cp * v * L / k
q1 = k * (T1 - T0) / L * (1.0 + 2.0 * np.exp(-((v/L) * (np.sqrt(R**2+(np.pi)**2)-R))*time))
t = time

R = 0.5 * rho * cp * v * L / k
v_over_L = v/L
q = np.zeros(len(t))
for i in range(len(t)):
    sigma = 0.0
    time = t[i]
    for n in range(1,10):
        sigma = sigma + np.exp(-((v/L)*(np.sqrt(R**2+(n*np.pi)**2)-R))*time)
    q[i] = k * (T1 - T0) / L * (1.0 + 2.0 * sigma)
    
print(len(q))
print(len(t))
```
